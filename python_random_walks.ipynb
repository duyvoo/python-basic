```python
import random
import matplotlib.pyplot as pl
import numpy  as np
```

```python
position = 0
walk = [position] #list
nsteps = 1000
for i in range(nsteps):
    step = 1 if random.randint(0, 1) else -1 
    position += step
    walk.append(position)
plt.plot(walk[:100])
```

```python

n_walks = 5       # Number of random walks
n_steps = 1000    # Number of steps per walk

# Initialize a list to store all the walks
all_walks = []

# Simulate multiple random walks
for walk_num in range(n_walks):
    start_position = 0
    current_walk = [start_position]  # Start each walk at position 0
    
    for _ in range(n_steps):
        step = 1 if random.randint(0, 1) else -1  # Random step (+1 or -1)
        start_position += step
        current_walk.append(start_position)
    
    all_walks.append(current_walk)

# Plot the first 100 steps of each walk
plt.figure(figsize=(10, 6))
for walk in all_walks:
    plt.plot(walk[:1000])
```
